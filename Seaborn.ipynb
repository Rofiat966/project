```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
movie_df = pd.read_csv('movie_data.csv')
```

```python
movie_df.head()
```

```python
sns.lineplot(x='title_year',y='duration',data=movie_df,err_style=None)
plt.show()
```

```python
sns.histplot(x='duration',data=movie_df,bins=50)
plt.show()
```

```python
sns.kdeplot(x='duration',data=movie_df)
plt.show()
```

```python
sns.histplot(x='duration',data=movie_df,kde=True)
plt.show()
```

```python
sns.boxplot(x='country',y='imdb_score',data=movie_df[(movie_df.country=='USA')|
                                                     (movie_df.country=='UK')|
                                                    (movie_df.country=='China')|
                                                    (movie_df.country=='Germany')])
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,6))

sns.boxplot(x='country',y='imdb_score',data=movie_df[(movie_df.country=='USA')|
                                                     (movie_df.country=='UK')|
                                                    (movie_df.country=='China')|
                                                    (movie_df.country=='Germany')], ax=ax[0])

sns.violinplot(x='country',y='imdb_score',data=movie_df[(movie_df.country=='USA')|
                                                     (movie_df.country=='UK')|
                                                    (movie_df.country=='China')|
                                                    (movie_df.country=='Germany')],ax=ax[1])
plt.show()
```

```python
languages = movie_df.language.apply(lambda x: 'other' if x not in ['English','French','Spanish'] else x)

to_plot = languages.value_counts()
```

```python
to_plot
```

```python
sns.barplot(y= to_plot, x=to_plot.index)
plt.show()
```

```python
sns.barplot(x= to_plot, y=to_plot.index)
plt.show()
```

```python
sns.lmplot(x='num_user_for_reviews',y='imdb_score',data=movie_df)
plt.show()
```

```python
sns.scatterplot(x='num_user_for_reviews',y='imdb_score',data=movie_df)
plt.show()
```
